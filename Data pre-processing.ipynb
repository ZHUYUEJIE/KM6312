# each_post dataset

```python
each_post=data.copy()
```

```python
each_post=each_post.drop('post_id',axis=1)
each_post=each_post.drop('user_id',axis=1)
#each_post=each_post.drop('followers',axis=1)
#each_post=each_post.drop('friends',axis=1)
#each_post=each_post.drop('favourites',axis=1)
#each_post=each_post.drop('statuses',axis=1)
```

```python
each_post
```

```python
each_post['post_created']
```

```python
Day_of_week=[]
for all_time in each_post['post_created']:
    Day_of_week.append(re.findall(r'(.+) [J|F|M|A|S|O|N|D]',str(all_time))[0])
each_post.loc[:,'Day_of_week'] = Day_of_week
each_post
```

```python
each_post.post_created=each_post.post_created.apply(pd.to_datetime)

Exact_data=[]
for all_time in each_post.post_created:
    Exact_data.append(re.findall(r'(.+) ',str(all_time))[0])
each_post.loc[:,'Exact_data'] = Exact_data

each_post['year']=each_post.post_created.dt.year
each_post['month']=each_post.post_created.dt.month
each_post['day']=each_post.post_created.dt.day
each_post['time']=each_post.post_created.dt.time

each_post.drop('post_created', axis=1, inplace=True)
```

```python
each_post=each_post.drop_duplicates()
```

```python
each_post
```

```python
each_post.to_csv('dataset/each_post.csv', index = True)
```
